```python
import numpy as np
import random as rm
import pandas as pd
PSA_df = pd.read_csv('PSA.csv').dropna(axis=1, how='all')
PSAOpenPrices = PSA_df['Open'].to_list()
PSAClosePrices = PSA_df['Close'].to_list()
```

```python
# The statespace
states = ["Interval 1","Interval 2","Interval 3"]

# Possible sequences of events
transitionName = [["11","12","13"],["21","22","23"],["31","32","33"]] #p_ij

# Probabilities matrix (transition matrix)
transitionMatrix = [[0.1597,0.6845,0.1558],[0.1597,0.6845,0.1558],[0.1597,0.6845,0.1558]] 
#also the steady state matrix

# A function that implements the Markov model to predicted the state/mood.
days=251 # number of entries plus 1
d=0
daysList=[d]
    # Choose the starting state
activityToday = "Interval 1"
    # This will store the sequence of states taken. So, this only has the starting state for now.
activityList = [activityToday]
i = 0
    # To calculate the probability of the activityList
prob = 1
while i != days:
    if activityToday == "Interval 1":
        change = np.random.choice(transitionName[0],replace=True,p=transitionMatrix[0])
        if change == "11":
            prob = prob * 0.1597
            activityList.append("11")
            d = d + 1
            daysList.append(d)
            pass
        elif change == "12":
            prob = prob * 0.6845
            activityToday = "Interval 2"
            activityList.append("12")
            d = d + 1
            daysList.append(d)
        else:
            prob = prob * 0.1558
            activityToday = "Interval 3"
            activityList.append("13")
            d = d + 1
            daysList.append(d)
    elif activityToday == "Interval 2":
        change = np.random.choice(transitionName[1],replace=True,p=transitionMatrix[1])
        if change == "22":
            prob = prob * 0.6845
            activityList.append("22")
            d = d + 1
            daysList.append(d)
            pass
        elif change == "21":
            prob = prob *0.1597
            activityToday = "Interval 1"
            activityList.append("21")
            d = d + 1
            daysList.append(d)
        else:
            prob = prob * 0.1558
            activityToday = "Interval 3"
            activityList.append("23")
            d = d + 1
            daysList.append(d)
    elif activityToday == "Interval 3":
        change = np.random.choice(transitionName[2],replace=True,p=transitionMatrix[2])
        if change == "33":
            prob = prob * 0.1558
            activityList.append("33")
            d = d + 1
            daysList.append(d)
            pass
        elif change == "32":
            prob = prob * 0.6845
            activityToday = "Interval 2"
            activityList.append("32")
            d = d + 1
            daysList.append(d)
        else:
            prob = prob * 0.1597
            activityToday = "Interval 1"
            activityList.append("31")
            d = d + 1
            daysList.append(d)
    i += 1  
```

```python
price=200.90 #Price of frist entry of open price
priceList=[price]
# this is to simulate the price of the stock moving up and down between the intervals
for i in activityList: 
    if i == "11":
        price=price - 2.1 # The range of one interval 
        priceList.append(price)
    elif i == "12":
        price=price + 4.2 # The range of two interval 
        priceList.append(price)
    elif i == "13":
        price=price + (4.2 + 2.1) # The range of three interval 
        priceList.append(price)
    elif i == "21": 
        price=price - 2.1  
        priceList.append(price)
    elif i == "22":
        price=price + 1.05 # The range of 1/2 interval 
        priceList.append(price)
    elif i == "23":
        price=price + 2.1
        priceList.append(price)
    elif i == "31":
        price=price - (8.4)
        priceList.append(price)    
    elif i == "32":
        price=price - 4.2
        priceList.append(price)
    elif i == "33":
        price=price + 2.1
        priceList.append(price)
    else:
        price=price+0
```
